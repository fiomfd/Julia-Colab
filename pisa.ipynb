```julia
using Plots
using CSV
using Dates
using DataFrames
using Downloads
```

```julia
data=DataFrame(CSV.File("./pisa.csv", header=false, delim=','))
```

```julia
mat=parse.(Float64,Array(data[2:6,2:6]))
```

```julia
plot([mat[:,1],mat[:,2],mat[:,3],mat[:,4],mat[:,5]],
    grid=false,
    linewidth=2,
    title="Average performance of 15-year-old students on mathematics \n data sourced by Our World in Data",
    titlefont=font(12),
    right_margin=Plots.Measures.Length(:mm, 10.0),
    left_margin=Plots.Measures.Length(:mm, 5.0),
    xlabel="year",
    xticks = ([1 2 3 4 5], [2009,2012,2015,2018,2022]),
    yaxis="scores",
    legendfont=font(9),
    label=["Finland" "Hong Kong" "Japan" "South Korea" "Singapore"],
    palette = :seaborn_bright,
    legend = :bottomleft)
```

```julia
savefig("./pisa.png")
```

```julia
imd=[1 5 3 4 1
    ;
    5 7 5 7 5
    ;
    13 11 8 11 15
    ;
    23 23 27 28 20
    ;
    34 31 34 35 38]
imd=imd'
R=51*ones(5,5)-imd
```

```julia
plot([R[:,3],R[:,2],R[:,5],R[:,4],R[:,1]],
    grid=false,
    linewidth=2,
    ylim=(-5,55),
    title="IMD World Competitiveness Ranking",
    titlefont=font(12),
    right_margin=Plots.Measures.Length(:mm, 10.0),
    left_margin=Plots.Measures.Length(:mm, 5.0),
    xlabel="year",
    xticks = ([1 2 3 4 5], [2020,2021,2022,2023,2024]),
    yticks = ([1 11 21 31 41 50], [50,40,30,20,10,1]),
    yaxis="rank",
    legendfont=font(9),
    label=["Finland" "Hong Kong" "Japan" "South Korea" "Singapore"],
    palette = :seaborn_bright,
    legend = :bottomleft)
```

```julia
savefig("./imd.png")
```
