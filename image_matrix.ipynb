## <font color="magenta">乱数で3つの行列を生成しRGB画像を作成する</font>

```julia
using LinearAlgebra
using LowRankApprox
using Formatting
using ImageFiltering
using Images
using Plots
using ImageMagick
using Colors
using TestImages
#using ImageView
using ImageTransformations
```

```julia
X=rand(80:255, 10, 15);
Y=rand(80:255, 10, 15);
Z=rand(80:255, 10, 15);
R1=zeros(3,10,15);
R1[1,:,:]=X/255;
R=colorview(RGB,R1);
G1=zeros(3,10,15);
G1[2,:,:]=Y/255;
G=colorview(RGB,G1);
B1=zeros(3,10,15);
B1[3,:,:]=Z/255;
B=colorview(RGB,B1);
W1=zeros(3,10,15);
W1[1,:,:]=X/255;
W1[2,:,:]=Y/255;
W1[3,:,:]=Z/255;
W=colorview(RGB,W1);
```

```julia
p1=plot(R, 
    xaxis=false, 
    xticks=false, 
    yaxis=false, 
    yticks=false, 
    grid=false, 
    title="Grayscale Image (RED)")

p2=plot(G, 
    xaxis=false, 
    xticks=false, 
    yaxis=false, 
    yticks=false, 
    grid=false, 
    title="Grayscale Image (GREEN)")

p3=plot(B, 
    xaxis=false, 
    xticks=false, 
    yaxis=false, 
    yticks=false, 
    grid=false, 
    title="Grayscale Image (BLUE)")

p4=plot(W, 
    xaxis=false, 
    xticks=false, 
    yaxis=false, 
    yticks=false, 
    grid=false, 
    title="RGB Image");
```

```julia
plot(p1, p2, p3, p4, 
     layout=(2,2), 
     size=(700,700), 
     margin=Plots.Measures.Length(:mm, 2.0))
```
