```julia
using Plots
using LaTeXStrings
```

```julia
x = -3pi:0.01:3pi
f = sin.(x)

# Define K
K=6

    taylor_sin = sum(((-1)^k) * (x .^ (2k+1)) / factorial(2k+1) for k in 0:K)
    plot(x, f, label="sin(x)",ylim=(-1.1,1.1),lw=2,
         xticks=([-3*pi,-2*pi,-pi,0,pi,2*pi,3*pi],[L"-3\pi",L"-2\pi",L"-\pi",0,L"\pi",L"2\pi",L"3\pi"]),
         yticks=([-1,0,1]))
    plot!(x, taylor_sin, label="Taylor (K=$K)",lw=2)
```
